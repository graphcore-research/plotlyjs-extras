This notebook shows resizable Plotly charts with legend highlight behavior (dim others + darken selected).

```python
import plotly.graph_objects as go
from plotly.colors import sample_colorscale
from plotlyjs_extras import show_resizable
```

```python
x = [1, 2, 3, 4, 5, 6, 7]

# Matplotlib-like sequential palette (Viridis) sampled from Plotly colorscale
base_colors = sample_colorscale("Viridis", [i / 7 for i in range(8)])

series = [
    {
        "name": "Line 1",
        "y": [1.1, 1.5, 1.9, 2.2, 2.6, 3.0, 3.2],
        "dash": "solid",
        "width": 3,
        "alpha": 0.95,
    },
    {
        "name": "Line 2",
        "y": [0.9, 1.2, 1.6, 2.0, 2.3, 2.6, 2.9],
        "dash": "dash",
        "width": 2.6,
        "alpha": 0.88,
    },
    {
        "name": "Line 3",
        "y": [0.8, 1.1, 1.4, 1.8, 2.0, 2.2, 2.5],
        "dash": "dot",
        "width": 2.4,
        "alpha": 0.80,
    },
    {
        "name": "Line 4",
        "y": [1.5, 1.8, 2.1, 2.4, 2.7, 2.9, 3.1],
        "dash": "dashdot",
        "width": 2.4,
        "alpha": 0.74,
    },
    {
        "name": "Line 5",
        "y": [2.2, 2.1, 2.0, 1.95, 1.9, 1.85, 1.8],
        "dash": "longdash",
        "width": 2.8,
        "alpha": 0.65,
    },
    {
        "name": "Line 6",
        "y": [2.5, 2.35, 2.2, 2.1, 2.0, 1.95, 1.9],
        "dash": "longdashdot",
        "width": 2.4,
        "alpha": 0.55,
    },
    {
        "name": "Line 7",
        "y": [1.8, 1.7, 1.6, 1.65, 1.75, 1.9, 2.05],
        "dash": "solid",
        "width": 1.8,
        "alpha": 0.42,
    },
    {
        "name": "Line 8",
        "y": [1.35, 1.5, 1.65, 1.8, 1.95, 2.1, 2.25],
        "dash": "dot",
        "width": 1.8,
        "alpha": 0.36,
    },
]

fig = go.Figure()
for i, row in enumerate(series):
    r, g, b = base_colors[i].replace("rgb(", "").replace(")", "").split(",")
    color = f"rgba({r.strip()}, {g.strip()}, {b.strip()}, {row['alpha']})"
    fig.add_trace(
        go.Scatter(
            x=x,
            y=row["y"],
            mode="lines+markers",
            name=row["name"],
            legendgroup=row["name"],
            line={"dash": row["dash"], "width": row["width"], "color": color},
            marker={"size": 6, "color": color},
        )
    )

fig.update_layout(
    title="Legend click to focus a line",
    template="plotly_white",
    yaxis_title="Value",
    xaxis_title="Step",
)

show_resizable(fig, height="520px", dim_opacity=0.12, active_darken=0.25)
```

## Tips

- Click a legend item once to dim other traces and darken the selected group.
- Click the same legend item again to restore all traces.
- Drag the chart's bottom-right corner to resize.
