# plotlyjs-extras quick demo

This notebook shows resizable Plotly charts with legend highlight behavior (dim others + darken selected).

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/graphcore-research/plotlyjs-extras/blob/main/examples/legend_highlight_demo.ipynb)

```python
# For Colab or first-time setup, install from GitHub
# %pip -q install git+https://github.com/graphcore-research/plotlyjs-extras.git
```

```python
import plotly.graph_objects as go
from plotlyjs_extras import show_resizable
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=[1, 2, 3, 4],
        y=[2, 1, 3, 2],
        mode="lines+markers",
        name="Series A",
        legendgroup="A",
    )
)
fig.add_trace(
    go.Scatter(
        x=[1, 2, 3, 4],
        y=[1, 3, 2, 4],
        mode="lines+markers",
        name="Series B",
        legendgroup="B",
    )
)
fig.add_trace(
    go.Bar(
        x=[1, 2, 3, 4],
        y=[1.5, 2.2, 2.5, 3.0],
        name="Bars",
        legendgroup="C",
        opacity=0.8,
    )
)

fig.update_layout(
    title="Legend click to focus a group",
    template="plotly_white",
)

show_resizable(fig, height="500px", dim_opacity=0.12, active_darken=0.25)
```

## Tips

- Click a legend item once to dim other traces and darken the selected group.
- Click the same legend item again to restore all traces.
- Drag the chart's bottom-right corner to resize.
